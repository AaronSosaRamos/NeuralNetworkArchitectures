#Shuffle the order of the tensor's elements

So that the inherent structure doesn't affect the learning

```python
tensor_not_shuffled = tf.constant([[[1,2],
                                    [3,4]],

                                   [[5,6],
                                   [7,8]]])

tensor_not_shuffled.ndim
```

```python
tensor_not_shuffled
```

```python
tf.random.shuffle(tensor_not_shuffled)
```

Operational seed only => It changes

```python
tf.random.shuffle(tensor_not_shuffled, seed=42)
```

Global seed and operational seed => It doesn't change

```python
tf.random.set_seed(42)
tf.random.shuffle(tensor_not_shuffled, seed=42)
```

Global seed (The same sequence for every re-run of the program if the operation seed is not set)

```python
tf.random.set_seed(1234)
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

```python
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

```python
print(tf.random.uniform([1], seed=1234))
print(tf.random.uniform([1], seed=1234))
```

```python
tf.random.set_seed(1234)

@tf.function
def f():
  a = tf.random.uniform([1])
  b = tf.random.uniform([1])
  return a, b

@tf.function
def g():
  a = tf.random.uniform([1])
  b = tf.random.uniform([1])
  return a, b

print(f())  # prints '(A1, A2)'
print(g())  # prints '(A1, A2)'
```

#Interactions with seeds:
1. If neither the global seed nor the operation seed is set: A randomly picked seed is used for this op.


```python
# Create a random tensor
random_tensor = tf.random.uniform(shape=(3, 3))

# Print the random tensor
print("Random Tensor:")
print(random_tensor)

# Running the TensorFlow operations again will produce different results due to the random nature
random_tensor_again = tf.random.uniform(shape=(3, 3))
print("\nRandom Tensor Again:")
print(random_tensor_again)
```

2. If the global seed is set, but the operation seed is not: The system deterministically picks an operation seed in conjunction with the global seed so that it gets a unique random sequence. Within the same version of tensorflow and user code, this sequence is deterministic. However across different versions, this sequence might change. If the code depends on particular seeds to work, specify both global and operation-level seeds explicitly.


```python
# Set the global seed
tf.random.set_seed(42)

# Create a random tensor
random_tensor = tf.random.uniform(shape=(3, 3))

# Print the random tensor
print("Random Tensor:")
print(random_tensor)

# Running the TensorFlow operations again will produce the same random tensor due to the specified seeds
random_tensor_again = tf.random.uniform(shape=(3, 3))
print("\nRandom Tensor Again:")
print(random_tensor_again)
```

3. If the operation seed is set, but the global seed is not set: A default global seed and the specified operation seed are used to determine the random sequence.


```python
# Create a random tensor
random_tensor = tf.random.uniform(shape=(3, 3), seed=123)

# Print the random tensor
print("Random Tensor:")
print(random_tensor)

# Running the TensorFlow operations again will produce the same random tensor due to the specified operation seed
random_tensor_again = tf.random.uniform(shape=(3, 3), seed=123)
print("\nRandom Tensor Again:")
print(random_tensor_again)
```

4. If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence.


```python
# Set the global seed
tf.random.set_seed(42)

# Create a random tensor
random_tensor = tf.random.uniform(shape=(3, 3), seed=13)

# Print the random tensor
print("Random Tensor:")
print(random_tensor)

# Running the TensorFlow operations again will produce the same random tensor due to the specified seeds
random_tensor_again = tf.random.uniform(shape=(3, 3), seed=13)
print("\nRandom Tensor Again:")
print(random_tensor_again)
```
